```python
#import the libraries needed
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import cufflinks as cf
import seaborn as sns
import plotly.express as px
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
#import data
titanic_data = pd.read_csv('./Titanic_Data.csv')
pd.set_option('max_column', None)
titanic_data.head()
```

## 1. Show the age distribution in the data using a histogram.

```python
age histogram
fig = px.histogram(titanic_data, x="Age")
fig.show()
```

## 2. Show the age distribution based on gender using a histogram

```python
#histogram age distribution based on gender
fig = px.histogram(titanic_data, x="Age", color="Sex" )
fig.show()
```

## 3. Using the function df.corr() to identify the correlation within the data, represent its results using a form of a matrix plot

```python
#correlation for the entire data
titanic_matrix = titanic_data.corr()
titanic_matrix
```

```python
fig = px.imshow(titanic_matrix)
fig.show()
```

## 4. Pivot the data setting the column Pclass as the columns and Fare as the values. From the resulting structure, use a boxplot to show the distribution of the values in its 3 columns.

```python
#pivot table
pclass_data = pd.pivot_table(titanic_data, index="PassengerId", columns="Pclass", values="Fare")
pclass_data
```

```python
#boxplot
fig = px.box(pclass_data)
fig.show()
```

## 5. Graph the value counts of the number of passengers who survived and did not survive based on gender using a stacked bar graph.

```python
fig = px.bar(titanic_data, x="Sex", y="Survived", barmode="stack", category_orders={"Survived": ["0", "1"]})
fig.show()
```

```python
survived = titanic_data.loc[titanic_data['Survived']==1, :]['Sex'].value_counts()
died     = titanic_data.loc[titanic_data['Survived']==0, :]['Sex'].value_counts()
survival  = pd.DataFrame([survived,died])
survival.index=['survived','died']
fig =px.bar(survival)
fig.show()
```

## 6. Using a scatter plot, plot the ages to the fare paid by the each passenger based on their gender.

```python
fig = px.scatter(titanic_data, x="Age", y="Fare", color ="Sex")
fig.show()
```

## 7. Plot a bubble plot of the ages to the fare paid by each passenger categorizing whether they survived or not. The size of each bubble should be determined by the passenger class and the name of each individual as the hover name.

```python
fig= px.scatter(titanic_data, x="Age", y="Fare", size='Pclass',color='Survived',hover_name='Name')
fig.show()
```
